```python
import pandas as pd
import matplotlib.pyplot as plt
# อ่านไฟล์ .csv
data = pd.read_csv("Data.csv")
# ดึงคอลัมน์
distance = data["distance"]
yaw_angle = data["yaw_angle"]
# พล็อตกราฟ
plt.figure(figsize=(10,5))
plt.plot(yaw_angle, distance, 'b-', marker='o')
plt.xlabel("Distance")
plt.ylabel("Yaw Angle (degrees)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Data.csv")
# ข้อมูลเซนเซอร์: ระยะทางและมุม yaw (ในองศา)
distance = data["distance"]
yaw_angle = data["yaw_angle"]
# แปลงมุมเป็นเรเดียน
yaw_angle = np.deg2rad(yaw_angle)
# คำนวณพิกัด x, y ของปลายเวกเตอร์
x = np.cos(yaw_angle) * distance
y = np.sin(yaw_angle) * distance
# วาดจุดปลายเวกเตอร์
plt.figure(figsize=(6,6))
plt.scatter(x, y, color='red', marker='o') # ใช้ scatter เพื่อแสดงจุดปลาย
# กำหนดขอบเขตของแกน X, Y
plt.xlim(-max(distance), max(distance))
plt.ylim(-max(distance), max(distance))
# ตั้งชื่อแกนและตาราง
plt.xlabel('X')
plt.ylabel('Y')
plt.xlabel("Distance")
plt.ylabel("Yaw Angle (degrees)")
plt.grid(True)
# ให้สัดส่วนแกน X, Y เท่ากัน
plt.gca().set_aspect('equal', adjustable='box')
# เพิ่มจุดเริ่มต้น (0,0) เพื่อให้เห็นจุดอ้างอิง
plt.scatter(0, 0, color='blue', marker='x', s=100, label='Start Point')
plt.legend()
plt.show()
```
