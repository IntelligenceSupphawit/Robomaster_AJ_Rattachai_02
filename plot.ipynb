```python
import pandas as pd
import matplotlib.pyplot as plt
# อ่านไฟล์ .csv
data = pd.read_csv("Data.csv")
# ดึงคอลัมน์
distance = data["distance"]
yaw_angle = data["yaw_angle"]
# พล็อตกราฟ
plt.figure(figsize=(10,5))
plt.plot(yaw_angle, distance, 'b-', marker='o')
plt.xlabel("Distance")
plt.ylabel("Yaw Angle (degrees)")
plt.grid(True)
plt.show()
```
